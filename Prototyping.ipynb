```python
import nibabel as nib
import matplotlib.pyplot as plt

sample_dir = "local_data/RSNA_ASNR_MICCAI_BraTS2021_ValidationData/BraTS2021_00719"
filenames = ["BraTS2021_00719_flair.nii.gz",  "BraTS2021_00719_t1ce.nii.gz",  "BraTS2021_00719_t1.nii.gz",  "BraTS2021_00719_t2.nii.gz"]
fig, ax = plt.subplots(2, 2, figsize=(15,15))
slice_to_show = 75  # 0-155
for i, fname in enumerate(filenames):
    slices = nib.load(f"{sample_dir}/{fname}").get_fdata()  # size is (240, 240, 155)
    ax[int(i/2)][i%2].imshow(slices[:,:,slice_to_show], "gray")
    ax[int(i/2)][i%2].set_xticks([])
    ax[int(i/2)][i%2].set_yticks([])
    ax[int(i/2)][i%2].set_xlabel(fname)
plt.show()
```
