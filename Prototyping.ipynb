```python
from pathlib import Path
import nibabel as nib
import matplotlib.pyplot as plt

sample_dir = "local_data/RSNA_ASNR_MICCAI_BraTS2021_TrainingData_16July2021/BraTS2021_00221"
fig, ax = plt.subplots(2,3, figsize=(15,15))
slice_to_show = 75  # 0-155
for i, fname in enumerate(Path(sample_dir).glob("*")):
    nifti = nib.load(fname)
    slices = nifti.get_fdata()  # size is (240, 240, 155)
    ax[int(i/3)][i%3].imshow(slices[:,:,slice_to_show], "gray")
    ax[int(i/3)][i%3].set_xticks([])
    ax[int(i/3)][i%3].set_yticks([])
    ax[int(i/3)][i%3].set_xlabel(fname.name)
plt.show()
```
